#### üìò Short Explanation  
This code creates data generators for loading and preprocessing the **Cats vs Dogs** images before model training.  

1. **Libraries:**  
   `ImageDataGenerator` is imported from `tensorflow.keras.preprocessing.image` to automatically read, resize, and normalize images.  

2. **Parameters:**  
   - `IMG_SIZE = (150, 150)` sets a fixed image size for all samples.  
   - `BATCH = 32` defines how many images are processed at once during training.  

3. **Data Generators:**  
   - `train_gen` reads images from the `train_dir` folder.  
   - `val_gen` reads images from the `val_dir` folder.  
   Both generators scale pixel values from 0‚Äì255 to 0‚Äì1 (`rescale=1./255`) and assign binary labels (cat/dog).  

4. **Output:**  
   The function prints how many images were found in each directory and confirms that data is ready for model training.  

---

#### üìò Qƒ±sa ƒ∞zah  
Bu kod **Cats vs Dogs** dataset-i √º√ß√ºn ≈ü…ôkill…ôri oxuyub emal ed…ôn data generator-lar yaradƒ±r.  

1. **Kitabxana:**  
   `ImageDataGenerator` ≈ü…ôkill…ôri avtomatik olaraq oxumaq, √∂l√ß√ºs√ºn√º d…ôyi≈üm…ôk v…ô normalizasiya etm…ôk √º√ß√ºn istifad…ô olunur.  

2. **Parametrl…ôr:**  
   - `IMG_SIZE = (150, 150)` ‚Üí b√ºt√ºn ≈ü…ôkill…ôr eyni √∂l√ß√ºy…ô g…ôtirilir.  
   - `BATCH = 32` ‚Üí model h…ôr addƒ±mda 32 ≈ü…ôkil √ºz…ôrind…ô t…ôlim aparƒ±r.  

3. **Generator-lar:**  
   - `train_gen` ‚Üí `train_dir` qovluƒüundakƒ± ≈ü…ôkill…ôri oxuyur.  
   - `val_gen` ‚Üí `val_dir` qovluƒüundakƒ± ≈ü…ôkill…ôri oxuyur.  
   H…ôr ikisi `rescale=1./255` il…ô piksell…ôri 0‚Äì1 aralƒ±ƒüƒ±na salƒ±r v…ô `class_mode="binary"` il…ô pi≈üik v…ô itl…ôri iki sinif kimi etiketl…ôyir.  

4. **N…ôtic…ô:**  
   Kod h…ôr iki qovluqda ne√ß…ô ≈ü…ôkil olduƒüunu √ßap edir v…ô m…ôlumatlarƒ±n model t…ôlimin…ô hazƒ±r olduƒüunu g√∂st…ôrir.


```python
from tensorflow.keras.preprocessing.image import ImageDataGenerator

IMG_SIZE = (150, 150)
BATCH = 32

train_gen = ImageDataGenerator(rescale=1./255).flow_from_directory(
    train_dir,
    target_size=IMG_SIZE,
    batch_size=BATCH,
    class_mode="binary"
)

val_gen = ImageDataGenerator(rescale=1./255).flow_from_directory(
    val_dir,
    target_size=IMG_SIZE,
    batch_size=BATCH,
    class_mode="binary"
)
```

#### üìò Short Explanation  
This cell builds and compiles a **Convolutional Neural Network (CNN)** model to classify cats and dogs.  

1. **Model structure:**  
   A `Sequential` model is created with multiple convolutional and pooling layers to extract features from images.  

2. **Layers overview:**  
   - `Conv2D + MaxPooling2D` layers learn visual features at increasing depth (32 ‚Üí 64 ‚Üí 128 filters).  
   - `Flatten` converts the 2D data into a 1D vector.  
   - `Dense(256, relu)` learns complex patterns, and `Dense(1, sigmoid)` outputs a binary result.  

3. **Compilation:**  
   The model uses the **Adam** optimizer, **binary crossentropy** loss, and tracks **accuracy** as a metric.  

4. **Summary:**  
   `model.summary()` displays the architecture and total parameters.  

---

#### üìò Qƒ±sa ƒ∞zah  
Bu cell **pi≈üik v…ô itl…ôrin t…ôsnifatƒ± √º√ß√ºn CNN modeli** qurur v…ô kompilyasiya edir.  

1. **Model qurulu≈üu:**  
   `Sequential` modeli qatlarƒ± ardƒ±cƒ±l ≈ü…ôkild…ô …ôlav…ô edir.  

2. **Qatlar:**  
   - `Conv2D + MaxPooling2D` ≈ü…ôkill…ôrd…ôn x√ºsusiyy…ôtl…ôr √ßƒ±xarƒ±r (filtrl…ôr: 32 ‚Üí 64 ‚Üí 128).  
   - `Flatten` veril…ônl…ôri bir √∂l√ß√ºl√º massiv…ô √ßevirir.  
   - `Dense(256, relu)` v…ô `Dense(1, sigmoid)` n…ôtic…ôni pi≈üik/it kimi t…ôsnif edir.  

3. **Kompilyasiya:**  
   `Adam` optimizatoru, `binary_crossentropy` z…ôr…ôr funksiyasƒ± v…ô `accuracy` metrikasƒ± il…ô model t…ôlim…ô hazƒ±rdƒ±r.  

4. **X√ºlas…ô:**  
   `model.summary()` modelin b√ºt√ºn qatlarƒ±nƒ± v…ô parametrl…ôrini g√∂st…ôrir.


```python
from tensorflow.keras import layers, models

IMG_SIZE = (150, 150)

model = models.Sequential([
    layers.Conv2D(32, (3,3), activation='relu', input_shape=(*IMG_SIZE, 3)),
    layers.MaxPooling2D(2,2),

    layers.Conv2D(64, (3,3), activation='relu'),
    layers.MaxPooling2D(2,2),

    layers.Conv2D(128, (3,3), activation='relu'),
    layers.MaxPooling2D(2,2),

    layers.Flatten(),
    layers.Dense(256, activation='relu'),
    layers.Dense(1, activation='sigmoid')  # binary output
])

model.compile(optimizer='adam', loss='binary_crossentropy', metrics=['accuracy'])
model.summary()
```

#### üìò Short Explanation  
This cell trains the CNN model for **5 epochs** and visualizes the training and validation performance.  

1. **Epochs:**  
   The model is trained for 5 full passes through the dataset (epochs).  

2. **Training process:**  
   `model.fit()` trains the model using `train_gen` and evaluates it on `val_gen` after each epoch.  
   The training history (accuracy and loss) is stored in the `history` variable.  

3. **Visualization:**  
   Two plots show how **accuracy** and **loss** evolve over time for both training and validation sets.  
   This helps identify overfitting or underfitting.  

---

#### üìò Qƒ±sa ƒ∞zah  
Bu cell **CNN modelini 5 epoch …ôrzind…ô √∂yr…ôdir** v…ô n…ôtic…ôl…ôri qrafikl…ôrl…ô g√∂st…ôrir.  

1. **Epochs:**  
   Model dataset √ºz…ôrind…ô 5 d…ôf…ô tam ke√ßid edir (sƒ±naq √º√ß√ºn kifay…ôtdir).  

2. **T…ôlim prosesi:**  
   `model.fit()` funksiyasƒ± `train_gen` m…ôlumatlarƒ±nƒ± t…ôlim √º√ß√ºn, `val_gen` m…ôlumatlarƒ±nƒ± is…ô doƒürulama √º√ß√ºn istifad…ô edir.  
   T…ôlim n…ôtic…ôl…ôri `history` d…ôyi≈ü…ônind…ô saxlanƒ±lƒ±r.  

3. **Vizualizasiya:**  
   Qrafikl…ôrd…ô **d…ôqiqlik (accuracy)** v…ô **itki (loss)** d…ôy…ôrl…ôrinin zamanla d…ôyi≈üm…ôsi g√∂st…ôrilir.  
   Bu, modelin d√ºzg√ºn √∂yr…ônib-√∂yr…ônm…ôdiyini v…ô overfitting olub-olmadƒ±ƒüƒ±nƒ± g√∂rm…ôy…ô k√∂m…ôk edir.


```python
EPOCHS = 5  # ilk sƒ±naq √º√ß√ºn 5 kifay…ôtdir

history = model.fit(
    train_gen,
    epochs=EPOCHS,
    validation_data=val_gen
)

# Accuracy v…ô loss qrafiki
import matplotlib.pyplot as plt

plt.figure(figsize=(10,4))
plt.subplot(1,2,1)
plt.plot(history.history['accuracy'], label='train_acc')
plt.plot(history.history['val_accuracy'], label='val_acc')
plt.title('Accuracy'); plt.legend()

plt.subplot(1,2,2)
plt.plot(history.history['loss'], label='train_loss')
plt.plot(history.history['val_loss'], label='val_loss')
plt.title('Loss'); plt.legend()
plt.tight_layout()
plt.show()
```

1. **Augmentasiya parametrl…ôri:**  
   - `rotation_range`, `width/height_shift`, `shear_range`, `zoom_range`, `horizontal_flip` ≈ü…ôkill…ôr…ô t…ôsad√ºfi d…ôyi≈üiklikl…ôr edir.  
   - `rescale=1./255` piksell…ôri normalla≈üdƒ±rƒ±r.  

2. **T…ôlim generatoru:**  
   `flow_from_directory()` ≈ü…ôkill…ôri oxuyur v…ô bu transformasiyalarƒ± h…ôr batch √º√ß√ºn t…ôtbiq edir.
